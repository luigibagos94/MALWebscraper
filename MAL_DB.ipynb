```python
import pandas as pd
import plotly.express as px
import streamlit as st
import numpy as np
import ast

# Specify the full file path including the file name and extension
file_path = '/Users/luigibagos/Desktop/MAL_Extract/MAL_DF.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

df
```

```python
df = df.dropna(subset=['Score', 'Members'])
df = df.reset_index(drop=True)
df['Demographic'].fillna('Unknown', inplace=True)
df['Start_Date'] = pd.to_datetime(df['Start_Date'])
df['Month'] = df['Start_Date'].dt.month
df['Total_Watchtime_Hr'] = round(df['Episodes'] * df['Min_Per_Ep'] / 60, 2)
df
```

```python
# Convert the strings to actual lists using ast.literal_eval, handling NaN values
def safe_eval(val):
    try:
        return ast.literal_eval(val)
    except (SyntaxError, ValueError):
        return val  # Return the value as is if it's not a valid literal

df['Genres'] = df['Genres'].apply(safe_eval)

# Explode the 'Genre' column to separate rows for each genre
separated_genres = df.explode('Genres')

# Count the number of occurrences of each genre
genre_counts = separated_genres['Genres'].value_counts()

# Convert the series genre_counts into a DataFrame
genre_counts_df = genre_counts.reset_index()
genre_counts_df.columns = ['Genres', 'Count']

genre_counts_df
```

```python
# Sample DataFrame
data = {'Year': [2019, 2019, 2019, 2020, 2020, 2020],
        'Members': [100, 100, 120, 130, 100, 100],
        'Score': [80, 90, 75, 85, 95, 92]}

df_selection = pd.DataFrame(data)

# Rank by 'Score' column and break ties using 'Members' column
df_selection['Combined'] = df_selection[['Members', 'Score']].apply(tuple, axis=1)
df_selection['Rank'] = df_selection.groupby('Year')['Combined'].rank(ascending=False).astype(int)
df_selection = df_selection.sort_values(by=['Year', 'Rank'])

df_selection
```
